This file covers the monthly time-shifting of the data from the first Pioneer probe. We tried to come up with a mathematical equation for what the time-shift should be, but ultimately it was more effective to simply plot the data and then shift it based off of the visual difference between the lines on the plots. This approach becomes less effective as there are more differences in terms of distance, longitude, and latitude.

```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt

fig = plt.figure()
fig.suptitle("Data From Pioneer 05/01/1972 to 06/01/1972")

df = pd.read_csv('csv files/Pioneer2.csv')
df['EPOCH_yyyy-mm-ddThh:mm:ss.sssZ'] = pd.to_datetime(df['EPOCH_yyyy-mm-ddThh:mm:ss.sssZ'])

df = df[df.PIONEER_DIST_Au != -1.0E+31]
df = df[df.PIONEER_LAT_deg != -1.0E+31]
df = df[df.PIONEER_LONG_deg != -1.0E+31]
df = df[df.PIONEER_BR_nT != -1.0E+31]
df = df[df.PIONEER_BT_nT != -1.0E+31]
df = df[df.PIONEER_BN_nT != -1.0E+31]
df = df[df.PIONEER_MAG_AVG_nT != -1.0E+31]
df = df[df.PIONEER_SPEED_kms != -1.0E+31]
df = df[df.PIONEER_DENSITY_Ncm3 != -1.0E+31]
df = df[df.PIONEER_TEMP_DegK != -1.0E+31]

#df = df.resample('H', on = 'EPOCH_yyyy-mm-ddThh:mm:ss.sssZ').mean().reset_index()

df.describe()

df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_DIST_Au')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_LAT_deg')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_LONG_deg')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_BR_nT')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_BT_nT')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_BN_nT')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_MAG_AVG_nT')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_SPEED_kms')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_DENSITY_Ncm3')
df.plot(kind='line', x='EPOCH_yyyy-mm-ddThh:mm:ss.sssZ', y='PIONEER_TEMP_DegK')

```
